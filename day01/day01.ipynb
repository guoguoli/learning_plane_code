```python
import numpy as np
import matplotlib.pyplot as plt

# 创建单位正方形顶点（二维向量集合）
square = np.array([[0, 1, 1, 0, 0],  # x坐标
                   [0, 0, 1, 1, 0]]) # y坐标

# 定义复合线性变换：先旋转45°，再非均匀缩放
theta = np.pi / 4  # 45度弧度制
rotation = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta), np.cos(theta)]])
scaling = np.array([[1.5, 0],   # x轴缩放1.5倍
                    [0, 0.8]])  # y轴缩放0.8倍

# 矩阵乘法表示变换复合：先旋转后缩放
transform = scaling @ rotation  # 注意矩阵乘法顺序

# 应用变换：矩阵乘以每个向量
transformed = transform @ square

# 可视化对比
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 原始正方形
axes[0].plot(square[0], square[1], 'b-', linewidth=2)
axes[0].fill(square[0], square[1], 'blue', alpha=0.2)
axes[0].set_title('原始单位正方形', fontsize=14)
axes[0].set_xlabel('x轴')
axes[0].set_ylabel('y轴')
axes[0].grid(True, alpha=0.3)
axes[0].axis('equal')

# 变换后的平行四边形
axes[1].plot(transformed[0], transformed[1], 'r-', linewidth=2)
axes[1].fill(transformed[0], transformed[1], 'red', alpha=0.2)
axes[1].set_title('线性变换后：旋转45°+非均匀缩放', fontsize=14)
axes[1].set_xlabel('x轴')
axes[1].set_ylabel('y轴')
axes[1].grid(True, alpha=0.3)
axes[1].axis('equal')

plt.tight_layout()
plt.show()

# 数学验证：变换矩阵的行列式值
det_value = np.linalg.det(transform)
print(f"变换矩阵的行列式值: {det_value:.4f}")
print(f"几何解释：面积变为原来的 {det_value:.2f} 倍")
```

```python
#numpy 高级线性代数实战
import numpy as np
print("🔢 NumPy高级线性代数实战")
print("=" * 60)

# 1. 矩阵乘法
# 矩阵乘法是线性代数中最常用的运算符，它将两个矩阵相乘，得到一个新的矩阵。
# 矩阵乘法的运算规则是：两个矩阵的行数等于第一个矩阵的列数，第二个矩阵的列数等于第二个矩阵的行数。
# 矩阵乘法的结果是一个新的矩阵，其行数等于第一个矩阵的行数，列数等于第二个矩阵的列数。
# 矩阵乘法的运算结果可以用一个公式来表示：
# C = A @ B
# 其中，C是乘积矩阵，A和B是两个矩阵。  
# 矩阵乘法的运算结果是一个新的矩阵，其元素是两个矩阵对应位置元素的乘积。
#代码示列
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
arr3 = arr1 @ arr2 
print(arr3)
# 2. 矩阵的转置
# 矩阵的转置是指矩阵的行列互换，即将矩阵的行变为列，将矩阵的列变为行。
# 矩阵的转置运算结果是一个新的矩阵，其行数等于原矩阵的列数，列数等于原矩阵的行数。
# 矩阵的转置运算可以用一个公式来表示：
# C = A.T
# 其中，C是转置矩阵，A是矩阵。
# 代码示列
arr4 = np.array([[1, 2], [3, 4]])
arr5 = arr4.T
print(arr5)
# 3. 矩阵的逆
# 矩阵的逆是指矩阵的乘法逆元，即矩阵乘以其转置矩阵，得到的矩阵是单位矩阵。
# 矩阵的逆运算结果是一个新的矩阵，其行数等于原矩阵的列数，列数等于原矩阵的行数。
# 矩阵的逆运算可以用一个公式来表示：
# C = np.linalg.inv(A)
# 其中，C是逆矩阵，A是矩阵。
# 代码示列
arr6 = np.array([[1, 2], [3, 4]])
arr7 = np.linalg.inv(arr6)
print(arr7)
# 4. 矩阵的行列式
# 矩阵的行列式是指矩阵的乘积的第n行第n列元素，其中n是矩阵的阶数。  
# 矩阵的行列式的值是矩阵的秩。
# 矩阵的行列式运算可以用一个公式来表示：
# det(A) = |A|
# 其中，det(A)是矩阵A的行列式，|A|是矩阵A的秩
# 代码示列
arr8 = np.array([[1, 2], [3, 4]])
arr9 = np.linalg.det(arr8)
print(arr9)
# 5. 矩阵的秩   
# 矩阵的秩是指矩阵的行列式的绝对值。
# 矩阵的秩的值是矩阵的行数与列数的最大值。
# 矩阵的秩运算可以用一个公式来表示：
# rank(A) = r
# 其中，rank(A)是矩阵A的秩，r是矩阵A的行数与列数的最大值。
# 代码示列
arr10 = np.array([[1, 2], [3, 4]])
arr11 = np.linalg.matrix_rank(arr10)
print(arr11)
# 6. 矩阵的特征值和特征向量
# 矩阵的特征值和特征向量是指矩阵的一些特殊的数值。
# 矩阵的特征值是指矩阵的一些特殊的数值，这些数值是矩阵的一些线性变换的结果。
# 矩阵的特征向量是指矩阵的特征值对应的单位向量。   
# 矩阵的特征值和特征向量运算可以用两个公式来表示：
# eigenvalues, eigenvectors = np.linalg.eig(A)
# 其中，eigenvalues是矩阵A的特征值，eigenvectors是矩阵A的特征向量。
# 代码示列
arr12 = np.array([[1, 2], [3, 4]])
arr13 = np.linalg.eig(arr12)
print(arr13)
# 7. 矩阵的条件数
# 矩阵的条件数是指矩阵的秩与矩阵的最大特征值的比值。
# 矩阵的条件数的值越大，矩阵越不稳定。
# 矩阵的条件数运算可以用一个公式来表示：
# cond(A) = ||A|| / ||A^2||
# 其中，cond(A)是矩阵A的条件数，||A||是矩阵A的秩，||A^2||是矩阵A的最大特征值的平方根。
# 代码示列
arr14 = np.array([[1, 2], [3, 4]])
arr15 = np.linalg.cond(arr14)
print(arr15)
#特殊矩阵的创建于性质验证
print("=" * 60)
print("🔢 特殊矩阵的创建与性质验证")
print("=" * 60)
# 1. 单位矩阵
# 单位矩阵是指对角线元素为1，其他元素为0的矩阵。
# 单位矩阵的创建可以用一个公式来表示：
# I = np.eye(n) 
#示列代码
i1 = np.eye(3)
print(i1)
# 2. 对称矩阵
# 对称矩阵是指对称的矩阵，即矩阵的转置等于矩阵本身。    
# 对称矩阵的创建可以用一个公式来表示：
# S = np.array([[a, b], [b, c]])
# 其中，S是对称矩阵，a, b, c是任意实数。
# 示列代码
s1 = np.array([[1, 2], [2, 1]])
print(s1)
# 3. 正定矩阵
# 正定矩阵是指矩阵的每个特征值都大于零的矩阵。
# 正定矩阵的创建可以用一个公式来表示：
# P = np.array([[a, b], [b, c]])
# 其中，P是正定矩阵，a, b, c是任意实数。
# 示列代码
p1 = np.array([[2, 1], [1, 2]])
print(p1)
# 4. 负定矩阵
# 负定矩阵是指矩阵的每个特征值都小于零的矩阵。
# 负定矩阵的创建可以用一个公式来表示：
# N = np.array([[a, b], [b, c]])
# 其中，N是负定矩阵，a, b, c是任意实数。
# 示列代码
n1 = np.array([[-2, 1], [1, -2]])
print(n1)
# 5. 奇异矩阵
# 奇异矩阵是指矩阵的行列式为0的矩阵。
# 奇异矩阵的创建可以用一个公式来表示：
# O = np.array([[a, b], [b, c]])
# 其中，O是奇异矩阵，a, b, c是任意实数。
# 示列代码
o1 = np.array([[1, 2], [2, 1]])
print(o1)
# 6. 行列式为0的矩阵
# 行列式为0的矩阵是指矩阵的行列式为0的矩阵。
# 行列式为0的矩阵的创建可以用一个公式来表示：
# Z = np.array([[a, b], [b, c]])
# 其中，Z是行列式为0的矩阵，a, b, c是任意实数。
# 示列代码
z1 = np.array([[1, 2], [2, 1]])
print(z1)
# 7. 秩为奇数的矩阵
# 秩为奇数的矩阵是指矩阵的秩为奇数的矩阵。
# 秩为奇数的矩阵的创建可以用一个公式来表示：
# R = np.array([[a, b], [b, c]])
# 其中，R是秩为奇数的矩阵，a, b, c是任意实数。
# 示列代码
r1 = np.array([[1, 2, 3], [2, 1, 4], [3, 4, 1]])
print(r1)
# 8. 秩为偶数的矩阵
# 秩为偶数的矩阵是指矩阵的秩为偶数的矩阵。
# 秩为偶数的矩阵的创建可以用一个公式来表示：
# E = np.array([[a, b], [b, c]])
# 其中，E是秩为偶数的矩阵，a, b, c是任意实数。
# 示列代码
e1 = np.array([[1, 2, 3], [2, 1, 4], [3, 4, 1]])
print(e1)  
# 9. 正交矩阵
# 正交矩阵是指矩阵的每一列都是单位向量的矩阵。
# 正交矩阵的创建可以用一个公式来表示：
# Q = np.array([[a, b], [b, c]])
# 其中，Q是正交矩阵，a, b, c是任意实数。
# 示列代码
q1 = np.array([[1, 0], [0, 1]])
print(q1)
np.random.seed(42)
a = np.random.rand(4, 4)
q,r = np.linalg.qr(a)
print(f"\n随机正交矩阵q (形状{q.shape}):")
print(f"前2行:\n{q[:2]}")
print(f"正交性验证:qᵀq≈ i1? (最大误差:{np.max(np.abs(q.T @ q - np.eye(4))):.6f})")
# 10. 酉矩阵
# 酉矩阵是指矩阵的每一列都是单位向量，且矩阵的转置等于其逆矩阵的矩阵。
# 酉矩阵的创建可以用一个公式来表示：
# H = np.array([[a, b], [b, c]])
# 其中，H是酉矩阵，a, b, c是任意实数。
# 示列代码
h1 = np.array([[1, 0], [0, 1]]) 
print(h1)
# 11. 矩阵分解的数学意义与实现
print("\n2. 矩阵分解的核心应用:")
matrix = np.array([[4, 2], [2, 3]])  # 对称正定矩阵
print(f"对称正定矩阵A:\n{matrix}")
# 特征值分解：A = VΛV⁻¹
eigenvalues, eigenvectors = np.linalg.eig(matrix)  # 特征值分解
print(f"\n特征值:\n{eigenvalues}")
print(f"特征向量矩阵 V:\n{eigenvectors}")
print(f"验证: ||AV - VΛ||_F = {np.linalg.norm(matrix @ eigenvectors - eigenvectors @ np.diag(eigenvalues)):.6f}")
# 奇异值分解：A = UΣVᵀ
u, s, vt = np.linalg.svd(matrix)  # 奇异值分解
print(f"\n奇异值:\n{s}")
print(f"左奇异矩阵 U:\n{u}")
print(f"右奇异矩阵 Vᵀ:\n{vt}")
print(f"Σ矩阵:\n{np.diag(s)}")
print(f"重构误差: ||A - UΣVᵀ||_F = {np.linalg.norm(matrix - u @ np.diag(s) @ vt):.6f}")
# 线性方程组求解的数值稳定性分析
print("\n3. 线性方程组求解的数值稳定性分析:")
a2 = np.array([[3,2],[1,-1]])
b2 = np.array([5,1])
cond_num = np.linalg.cond(a2)
print(f"系数矩阵条件数: κ(A) = {cond_num:.4f}")
print(f"解释: 条件数较大(>10³)时，小扰动可能导致解大幅变化")
x = np.linalg.solve(a2, b2)
print(f"方程组的解: x = {x[0]:.2f}, y = {x[1]:.2f}")
print(f"残差范数: ||Ax - b||₂ = {np.linalg.norm(a2 @ x - b2):.6f}")
```

```python
 #Pandas数据处理实战
import pandas as pd

np.random.seed(42)
schools_info = {
        '成都七中': {'district': '锦江区', 'tier': '顶尖', 'student_count': 3000},
        '树德中学': {'district': '青羊区', 'tier': '顶尖', 'student_count': 2800},
        '石室中学': {'district': '青羊区', 'tier': '顶尖', 'student_count': 3200},
        '成都外国语学校': {'district': '高新区', 'tier': '一流', 'student_count': 2500},
        '玉林中学': {'district': '武侯区', 'tier': '一流', 'student_count': 2000}
    }
n_students = 1000
schools = list(schools_info.keys())
districts = [schools_info[school]['district'] for school in schools]
tiers = [schools_info[school]['tier'] for school in schools]
student_counts = [schools_info[school]['student_count'] for school in schools]

school_probs = np.array(student_counts) / np.sum(student_counts)
print(school_probs)
print(f"1. 成都教育资源分布:")
for school, info in schools_info.items():
    print(f"     - {school}: {info['district']}区， {info['tier']}，约{info['student_count']}名学生")

data = {
    '学生ID':range(1,n_students+1),
    '姓名':[f'学生_{i:04d}' for i in range(1,n_students+1)],
    '性别':np.random.choice(['男', '女'], n_students,p=[0.52, 0.48]),
    '年龄':np.random.choice([15,16,17,18],n_students,p=[0.25,0.35,0.30,0.10]),

    '学校':np.random.choice(schools, n_students, p=school_probs),
    '所在区':np.random.choice(districts, n_students),
    '学校层级':np.random.choice(tiers, n_students),

    '语文':np.zeros(n_students),
    '数学':np.zeros(n_students),
    '英语':np.zeros(n_students),
    '物理':np.zeros(n_students),
    '化学':np.zeros(n_students),

    '每周学习时长(小时)':np.zeros(n_students),
    '每周辅导班时长(小时)':np.zeros(n_students),
    '教育APP使用指数':np.zeros(n_students),

    '学习压力指数': np.zeros(n_students),
    '睡眠质量评分': np.zeros(n_students),
    '课外活动参与度': np.zeros(n_students),
    
}
for i, school in enumerate(schools):
    mask = data['学校'] == school
    n_school_students = np.sum(mask)
    if n_school_students == 0:
        continue

    if schools_info[school]['tier'] == '顶尖':
        base_mean = 75
        base_std = 10
        pressure_mean = 70
    else:
        base_mean = 65
        base_std = 15
        pressure_mean = 60
    # 生成相关学科成绩
    # 语文-数学相关系数约0.6，数学-物理约0.7
    cov_matrix = np.array([
        [1.0, 0.6, 0.5, 0.4, 0.3],  # 语文
        [0.6, 1.0, 0.7, 0.6, 0.5],  # 数学
        [0.5, 0.7, 1.0, 0.5, 0.4],  # 英语
        [0.4, 0.6, 0.5, 1.0, 0.8],  # 物理
        [0.3, 0.5, 0.4, 0.8, 1.0]   # 化学
    ])
    # 生成多元正态分布成绩
    means = np.array([base_mean,base_mean+3,base_mean-2,base_mean-5,base_mean-7])
    scores = np.random.multivariate_normal(means, cov_matrix * base_std**2, n_school_students)
    data['语文'][mask] = scores[:,0].clip(0,100)
    data['数学'][mask] = scores[:,1].clip(0,100)
    data['英语'][mask] = scores[:,2].clip(0,100)
    data['物理'][mask] = scores[:,3].clip(0,100)
    data['化学'][mask] = scores[:,4].clip(0,100)
    # 生成行为数据（与成绩相关）
    data['每周学习时长(小时)'][mask] = np.random.normal(30, 8, n_school_students).clip(10, 50)
    data['每周辅导班时长(小时)'][mask] = np.random.normal(8, 4, n_school_students).clip(0, 20)
    data['教育APP使用指数'][mask] = np.random.normal(7, 3, n_school_students).clip(0, 15)

    # 心理指标
    data['学习压力指数'][mask] = np.random.normal(pressure_mean, 15, n_school_students).clip(20, 100)
    data['睡眠质量评分'][mask] = np.random.normal(70, 12, n_school_students).clip(30, 100)
    data['课外活动参与度'][mask] = np.random.normal(60, 20, n_school_students).clip(0, 100)
df = pd.DataFrame(data)
print(f"\n2. 数据集概览:")
print(f"   总记录数: {df.shape[0]}名学生")
print(f"   特征维度: {df.shape[1]}个字段")
print(f"\n3. 数据结构:")
print(df.dtypes.to_string())
print(f"\n4. 描述性统计（学科成绩）:")
score_columns = ['语文', '数学', '英语', '物理', '化学']
print(df[score_columns].describe().round(2))    
print(f"\n5. 学校分布:")
school_dist = df['学校'].value_counts()

for school, count in school_dist.items():
    tier = schools_info[school]['tier']
    district = schools_info[school]['district']
    print(f"   - {school}（{district}，{tier}）: {count}人 ({count/n_students*100:.1f}%)")

# 数据清洗与特征工程
print(f"\n6. 数据清洗与特征工程:")
 # 检查缺失值
missing_values = df.isnull().sum()
missing_cols = missing_values[missing_values > 0]
if len(missing_cols) > 0:
    print(f"   缺失值统计:\n{missing_cols}")
else:
    print("   无缺失值 ✓")
# 处理异常值（学科成绩限制在0-100）
for col in score_columns:
    df[col] = df[col].clip(0, 100)
print("   学科成绩异常值处理完成（限制在0-100范围内）")

# 创建衍生特征
print(f"\n7. 衍生特征创建:")
# 学科总分与平均分
df['总分'] = df[score_columns].sum(axis=1)
df['平均分'] = df[score_columns].mean(axis=1)

# 学科优势（最高分学科）
df['优势学科'] = df[score_columns].idxmax(axis=1)

# 文理倾向指数（文科=语文+英语，理科=数学+物理+化学）
df['文科得分'] = df['语文'] + df['英语']
df['理科得分'] = df['数学'] + df['物理'] + df['化学']
df['文理倾向指数'] = (df['文科得分'] - df['理科得分']) / 500  # 归一化到[-1,1]
    
# 学习效率指数 = 平均分 / sqrt(学习时长+辅导班时长)
total_study_hours = df['每周学习时长(小时)'] + df['每周辅导班时长(小时)']
df['学习效率指数'] = df['平均分'] / np.sqrt(total_study_hours)
    
# 综合素质评分（成绩60% + 课外活动20% + 睡眠质量20%）
df['综合素质评分'] = (df['平均分'] * 0.6 + 
                    df['课外活动参与度'] * 0.2 + 
                    df['睡眠质量评分'] * 0.2)
    
print("   衍生特征创建完成，新增字段:")
new_features = ['总分', '平均分',  '文科得分', '理科得分', 
                '文理倾向指数', '学习效率指数', '综合素质评分']
for feat in new_features:
    print(f"     - {feat}: {df[feat].describe()['mean']:.2f} ± {df[feat].describe()['std']:.2f}")
  
```

```python
#挑战1 线性代数综合应用 - 学生能力PCA分析【1】
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def pca_student_ability_analysis(df, score_columns):
    """
    成都中学生五科成绩PCA能力分析
    """
    plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB']
    plt.rcParams['axes.unicode_minus'] = False

    # ======================
    # 1. 数据预处理：标准化
    # ======================
    scores = df[score_columns].copy()
    scaler = StandardScaler()
    scores_scaled = scaler.fit_transform(scores)

    # ======================
    # 2. PCA建模（2主成分）
    # ======================
    pca = PCA(n_components=2, random_state=42)
    pca_result = pca.fit_transform(scores_scaled)

    explained_variance = pca.explained_variance_ratio_
    components = pca.components_

    # ======================
    # 3. 结果解读（成都教育特色）
    # ======================
    pc1_weights = pd.Series(components[0], index=score_columns)
    pc2_weights = pd.Series(components[1], index=score_columns)

    primary_factor = "第一主成分：综合学业基础能力（所有学科均正向贡献，数理学科权重更高）"
    secondary_factor = "第二主成分：文理科偏向能力（文科正、理科负 或 反之）"

    cor_matrix = df[score_columns].corr()
    subject_correlation = {
        "强相关": "数学-物理、物理-化学（理科高度相关）",
        "中等相关": "语文-英语（文科相关）",
        "弱相关": "文科与理科之间"
    }

    recommendations = [
        {
            "建议": "强化理科协同教学，数学、物理、化学同步提升",
            "数据依据": "数理化高度相关，PCA显示同属一个能力维度",
            "预期效果": "提升成都学生理科竞争力"
        },
        {
            "建议": "针对文理科偏向学生实施分层教学",
            "数据依据": "第二主成分体现文理分化明显",
            "预期效果": "减少偏科，促进均衡发展"
        }
    ]

    insights = {
        "primary_factor": primary_factor,
        "secondary_factor": secondary_factor,
        "subject_correlation": subject_correlation,
        "recommendations": recommendations
    }

    # ======================
    # 4. 可视化（3张图）
    # ======================
    fig = plt.figure(figsize=(18, 5))

    # ----------------------
    # 图1：PCA降维散点图
    # ----------------------
    ax1 = plt.subplot(1, 3, 1)
    colors = {'锦江区': 'red', '青羊区': 'blue', '高新区': 'green', '武侯区': 'orange'}
    for district, color in colors.items():
        mask = df['所在区'] == district
        ax1.scatter(
            pca_result[mask, 0],
            pca_result[mask, 1],
            s=20, alpha=0.6,
            color=color, label=district
        )
    ax1.set_xlabel(f'PC1（{explained_variance[0]:.1%}）')
    ax1.set_ylabel(f'PC2（{explained_variance[1]:.1%}）')
    ax1.set_title('成都中学生成绩PCA降维散点图')
    ax1.legend()
    ax1.grid(alpha=0.3)

    # ----------------------
    # 图2：方差解释率条形图
    # ----------------------
    ax2 = plt.subplot(1, 3, 2)
    ax2.bar(['PC1', 'PC2'], explained_variance, color=['#FF6B6B', '#4ECDC4'])
    ax2.set_title('方差解释率')
    ax2.set_ylabel('解释比例')
    for i, v in enumerate(explained_variance):
        ax2.text(i, v + 0.01, f'{v:.1%}', ha='center')

    # ----------------------
    # 图3：特征向量热力图
    # ----------------------
    ax3 = plt.subplot(1, 3, 3)
    im = ax3.imshow(components, cmap='RdBu_r', aspect='auto')
    ax3.set_xticks(range(len(score_columns)))
    ax3.set_yticks([0, 1])
    ax3.set_xticklabels(score_columns)
    ax3.set_yticklabels(['PC1', 'PC2'])
    ax3.set_title('学科-主成分贡献热力图')
    for i in range(2):
        for j in range(len(score_columns)):
            ax3.text(j, i, f'{components[i, j]:.2f}', ha='center', va='center', color='white', fontweight='bold')

    plt.tight_layout()
    plt.show()

    # ======================
    # 输出（严格按要求）
    # ======================
    return pca, explained_variance, components, insights


# ======================
# 调用示例（直接运行）
# ======================
pca_model, explained_variance, components, insights = pca_student_ability_analysis(df, score_columns)

# ======================
# 输出结果
# ======================
print("="*60)
print("📊 成都中学生成绩PCA综合分析报告")
print("="*60)
print(f"方差解释率：PC1 = {explained_variance[0]:.1%}，PC2 = {explained_variance[1]:.1%}")
print(f"累计解释率：{sum(explained_variance):.1%}\n")

print("🔍 主成分解读：")
print(insights['primary_factor'])
print(insights['secondary_factor'])

print("\n🔗 学科相关性：")
print(insights['subject_correlation']['强相关'])
print(insights['subject_correlation']['中等相关'])

print("\n🎯 成都教育改进建议：")
for i, rec in enumerate(insights['recommendations'], 1):
    print(f"{i}. {rec['建议']}")
    print(f"   数据依据：{rec['数据依据']}")
    print(f"   预期效果：{rec['预期效果']}\n")
```

```python
#挑战1 线性代数综合应用 - 学生能力PCA分析【2】
def create_chengdu_education_dataset():
    """
    创建模拟的成都教育科技数据集（增强版）
    
    数据特点：
    1. 成都真实中学名称和行政区划
    2. 学科成绩模拟（考虑正态分布和相关性）
    3. 学习行为与心理指标
    4. 成都教育特色字段
    
    返回：
    tuple: (DataFrame, 学科列名列表)
    """
    print("\n📊 第二部分：Pandas数据处理实战 - 成都教育科技数据集（增强版）")
    print("-" * 70)
    
    np.random.seed(42)  # 确保可复现
    
    # 成都教育资源分布（基于真实情况）
    schools_info = {
        '成都七中': {'district': '锦江区', 'tier': '顶尖', 'student_count': 3000},
        '树德中学': {'district': '青羊区', 'tier': '顶尖', 'student_count': 2800},
        '石室中学': {'district': '青羊区', 'tier': '顶尖', 'student_count': 3200},
        '成都外国语学校': {'district': '高新区', 'tier': '一流', 'student_count': 2500},
        '玉林中学': {'district': '武侯区', 'tier': '一流', 'student_count': 2000}
    }
    
    # 生成1000条学生记录
    n_students = 1000
    schools = list(schools_info.keys())
    districts = [schools_info[school]['district'] for school in schools]
    tiers = [schools_info[school]['tier'] for school in schools]
    
    # 按学校层级分配学生比例
    school_probs = [schools_info[school]['student_count'] for school in schools]
    school_probs = np.array(school_probs) / sum(school_probs)
    
    print(f"1. 成都教育资源分布:")
    for school, info in schools_info.items():
        print(f"   - {school}: {info['district']}区，{info['tier']}，约{info['student_count']}名学生")
    
    # 生成学生数据
    data = {
        '学生ID': range(1, n_students + 1),
        '姓名': [f'学生_{i:04d}' for i in range(1, n_students + 1)],
        '性别': np.random.choice(['男', '女'], n_students, p=[0.52, 0.48]),  # 成都中学生性别比例
        '年龄': np.random.choice([15, 16, 17, 18], n_students, p=[0.25, 0.35, 0.30, 0.10]),
        
        # 学校分配（考虑学校层级权重）
        '学校': np.random.choice(schools, n_students, p=school_probs),
        '所在区': np.random.choice(districts, n_students),
        '学校层级': np.random.choice(tiers, n_students),
        
        # 学科成绩生成（考虑学科相关性和学校差异）
        '语文': np.zeros(n_students),
        '数学': np.zeros(n_students),
        '英语': np.zeros(n_students),
        '物理': np.zeros(n_students),
        '化学': np.zeros(n_students),
        
        # 学习行为数据（成都特色）
        '每周学习时长(小时)': np.zeros(n_students),
        '每周辅导班时长(小时)': np.zeros(n_students),  # 成都补习文化
        '教育APP使用指数': np.zeros(n_students),  # 成都教育科技渗透率
        
        # 心理与健康指标
        '学习压力指数': np.zeros(n_students),
        '睡眠质量评分': np.zeros(n_students),
        '课外活动参与度': np.zeros(n_students),  # 成都素质教育特色
    }
    
    # 按学校层级生成差异化的成绩分布
    for i, school in enumerate(schools):
        mask = data['学校'] == school
        n_school_students = np.sum(mask)
        
        if n_school_students == 0:
            continue
            
        # 学校基准水平
        if schools_info[school]['tier'] == '顶尖':
            base_mean = 75
            base_std = 10
            pressure_mean = 70  # 顶尖学校压力较大
        else:
            base_mean = 65
            base_std = 15
            pressure_mean = 60
        
        # 生成相关学科成绩
        # 语文-数学相关系数约0.6，数学-物理约0.7
        cov_matrix = np.array([
            [1.0, 0.6, 0.5, 0.4, 0.3],  # 语文
            [0.6, 1.0, 0.7, 0.6, 0.5],  # 数学
            [0.5, 0.7, 1.0, 0.5, 0.4],  # 英语
            [0.4, 0.6, 0.5, 1.0, 0.8],  # 物理
            [0.3, 0.5, 0.4, 0.8, 1.0]   # 化学
        ])
        
        # 生成多元正态分布成绩
        means = np.array([base_mean, base_mean+3, base_mean-2, base_mean-5, base_mean-7])
        scores = np.random.multivariate_normal(means, cov_matrix * base_std**2, n_school_students)
        
        # 填充数据
        data['语文'][mask] = scores[:, 0].clip(0, 100)
        data['数学'][mask] = scores[:, 1].clip(0, 100)
        data['英语'][mask] = scores[:, 2].clip(0, 100)
        data['物理'][mask] = scores[:, 3].clip(0, 100)
        data['化学'][mask] = scores[:, 4].clip(0, 100)
        
        # 生成行为数据（与成绩相关）
        data['每周学习时长(小时)'][mask] = np.random.normal(30, 8, n_school_students).clip(10, 50)
        data['每周辅导班时长(小时)'][mask] = np.random.normal(8, 4, n_school_students).clip(0, 20)
        data['教育APP使用指数'][mask] = np.random.normal(7, 3, n_school_students).clip(0, 15)
        
        # 心理指标
        data['学习压力指数'][mask] = np.random.normal(pressure_mean, 15, n_school_students).clip(20, 100)
        data['睡眠质量评分'][mask] = np.random.normal(70, 12, n_school_students).clip(30, 100)
        data['课外活动参与度'][mask] = np.random.normal(60, 20, n_school_students).clip(0, 100)
    
    # 创建DataFrame
    df = pd.DataFrame(data)
    
    # 数据探索
    print(f"\n2. 数据集概览:")
    print(f"   总记录数: {df.shape[0]}名学生")
    print(f"   特征维度: {df.shape[1]}个字段")
    
    print(f"\n3. 数据结构:")
    print(df.dtypes.to_string())
    
    print(f"\n4. 描述性统计（学科成绩）:")
    score_columns = ['语文', '数学', '英语', '物理', '化学']
    print(df[score_columns].describe().round(2))
    
    print(f"\n5. 学校分布:")
    school_dist = df['学校'].value_counts()
    for school, count in school_dist.items():
        tier = schools_info[school]['tier']
        district = schools_info[school]['district']
        print(f"   - {school}（{district}，{tier}）: {count}人 ({count/n_students*100:.1f}%)")
    
    # 数据清洗与特征工程
    print(f"\n6. 数据清洗与特征工程:")
    
    # 检查缺失值
    missing_values = df.isnull().sum()
    missing_cols = missing_values[missing_values > 0]
    if len(missing_cols) > 0:
        print(f"   缺失值统计:\n{missing_cols}")
    else:
        print("   无缺失值 ✓")
    
    # 处理异常值（学科成绩限制在0-100）
    for col in score_columns:
        df[col] = df[col].clip(0, 100)
    print("   学科成绩异常值处理完成（限制在0-100范围内）")
    
    # 创建衍生特征
    print(f"\n7. 衍生特征创建:")
    
    # 学科总分与平均分
    df['总分'] = df[score_columns].sum(axis=1)
    df['平均分'] = df[score_columns].mean(axis=1)
    
    # 学科优势（最高分学科）
    df['优势学科'] = df[score_columns].idxmax(axis=1)
    
    # 文理倾向指数（文科=语文+英语，理科=数学+物理+化学）
    df['文科得分'] = df['语文'] + df['英语']
    df['理科得分'] = df['数学'] + df['物理'] + df['化学']
    df['文理倾向指数'] = (df['文科得分'] - df['理科得分']) / 500  # 归一化到[-1,1]
    
    # 学习效率指数 = 平均分 / sqrt(学习时长+辅导班时长)
    total_study_hours = df['每周学习时长(小时)'] + df['每周辅导班时长(小时)']
    df['学习效率指数'] = df['平均分'] / np.sqrt(total_study_hours)
    
    # 综合素质评分（成绩60% + 课外活动20% + 睡眠质量20%）
    df['综合素质评分'] = (df['平均分'] * 0.6 + 
                        df['课外活动参与度'] * 0.2 + 
                        df['睡眠质量评分'] * 0.2)
    
    print("   衍生特征创建完成，新增字段:")
    new_features = ['总分', '平均分', '文科得分', '理科得分', 
                   '文理倾向指数', '学习效率指数', '综合素质评分']
    for feat in new_features:
        print(f"     - {feat}: {df[feat].describe()['mean']:.2f} ± {df[feat].describe()['std']:.2f}")
    
    return df, score_columns, schools_info
def challenge_1_pca_framework(df, score_columns):
    """
    挑战1：学生能力PCA分析详细框架
    
    输入：
    df: 学生数据DataFrame
    score_columns: 学科列名列表
    
    输出：
    dict: PCA分析结果和可视化
    """
    print("\n🎯 挑战1：学生能力PCA分析详细框架")
    print("-" * 60)
    
    print("任务描述:")
    print("1. 对五科成绩进行PCA降维，提取2个主成分")
    print("2. 分析各主成分的学科权重和实际意义") 
    print("3. 可视化降维结果和教育洞察")
    print("4. 提出具体的教育改进建议")
    
    print("\n详细步骤:")
    
    steps = [
        "# 步骤1: 数据准备",
        "X = df[score_columns].values  # 提取成绩矩阵",
        "",
        "# 步骤2: 数据标准化（PCA前必须）",
        "from sklearn.preprocessing import StandardScaler",
        "scaler = StandardScaler()",
        "X_scaled = scaler.fit_transform(X)",
        "",
        "# 步骤3: PCA建模",
        "from sklearn.decomposition import PCA",
        "pca = PCA(n_components=2)",
        "X_pca = pca.fit_transform(X_scaled)",
        "",
        "# 步骤4: 结果分析",
        "print(f'方差解释率: {pca.explained_variance_ratio_}')",
        "print(f'累计解释率: {np.sum(pca.explained_variance_ratio_):.3f}')",
        "print(f'特征向量矩阵:\\n{pca.components_}')",
        "",
        "# 步骤5: 可视化（提供完整代码框架）",
        "fig, axes = plt.subplots(2, 3, figsize=(15, 10))",
        "",
        "# 5.1 PCA降维散点图",
        "scatter = axes[0, 0].scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.6, s=30)",
        "axes[0, 0].set_xlabel('第一主成分 (解释率: {:.1%})'.format(pca.explained_variance_ratio_[0]))",
        "axes[0, 0].set_ylabel('第二主成分 (解释率: {:.1%})'.format(pca.explained_variance_ratio_[1]))",
        "axes[0, 0].set_title('PCA降维结果 - 学生分布')",
        "axes[0, 0].grid(alpha=0.3)",
        "",
        "# 5.2 方差解释率条形图",
        "axes[0, 1].bar(range(1, 3), pca.explained_variance_ratio_, color='skyblue')",
        "axes[0, 1].set_xlabel('主成分')",
        "axes[0, 1].set_ylabel('方差解释率')",
        "axes[0, 1].set_title('各主成分方差解释率')",
        "for i, v in enumerate(pca.explained_variance_ratio_):",
        "    axes[0, 1].text(i+1, v+0.01, f'{v:.1%}', ha='center')",
        "",
        "# 5.3 特征向量热力图",
        "im = axes[0, 2].imshow(pca.components_, cmap='RdYlBu', aspect='auto')",
        "axes[0, 2].set_xticks(range(len(score_columns)))",
        "axes[0, 2].set_xticklabels(score_columns, rotation=45)",
        "axes[0, 2].set_yticks([0, 1])",
        "axes[0, 2].set_yticklabels(['PC1', 'PC2'])",
        "axes[0, 2].set_title('主成分与学科关系热图')",
        "plt.colorbar(im, ax=axes[0, 2])",
        "",
        "# 步骤6: 教育洞察与建议",
        "# 请基于PCA结果，回答以下问题并提出建议:",
        "# 1. 第一主成分主要反映学生的什么能力？",
        "# 2. 第二主成分补充了哪些信息？",
        "# 3. 哪些学科相关性最强？这对教学安排有什么启示？",
        "# 4. 基于成都教育现状，提出2条具体的改进建议",
    ]
    
    for step in steps:
        print(step)
    
    print("\n输入数据规格:")
    print(f"- 数据形状: {df.shape}")
    print(f"- 学科列: {score_columns}")
    print(f"- 样本数: {len(df)}")
    
    print("\n预期输出:")
    expected_output = {
        'pca_model': 'sklearn.decomposition.PCA对象',
        'explained_variance_ratio': 'array([PC1解释率, PC2解释率])',
        'components': 'array([[PC1各学科权重], [PC2各学科权重]])',
        'pca_scores': 'array([[学生1的PC1得分, PC2得分], ...])',
        'visualizations': ['散点图', '方差解释率图', '热力图'],
        'insights': {
            'primary_factor': '第一主成分解释',
            'secondary_factor': '第二主成分解释',
            'subject_correlations': '学科相关性分析',
            'recommendations': '2条教育建议（每条需包含数据依据）'
        }
    }
    
    for key, value in expected_output.items():
        print(f"- {key}: {value}")
    
    return expected_output

def challenge_1_pca_framework_new(df, score_columns):
    """
    挑战1：线性代数综合应用 - 学生能力PCA分析
    适配 create_chengdu_education_dataset() 返回的数据
    """
    # --------------------------
    # 中文显示设置
    # --------------------------
    plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB']
    plt.rcParams['axes.unicode_minus'] = False

    # ==========================
    # 1. 数据预处理：Z-score标准化
    # ==========================
    X = df[score_columns].copy()
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # ==========================
    # 2. PCA建模（2个主成分）
    # ==========================
    pca = PCA(n_components=2, random_state=42)
    pca_result = pca.fit_transform(X_scaled)

    explained_variance = pca.explained_variance_ratio_
    components = pca.components_

    # ==========================
    # 3. 主成分解读（成都教育分析）
    # ==========================
    pc1 = pd.Series(components[0], index=score_columns)
    pc2 = pd.Series(components[1], index=score_columns)

    primary_factor = "第一主成分：综合学业能力（所有学科正贡献，数理权重更高）"
    secondary_factor = "第二主成分：文理科偏向能力（文科正/理科负 或 相反）"

    subject_correlation = {
        "强相关": "数学-物理、物理-化学（理科高度相关）",
        "中等相关": "语文-英语（文科相关）",
        "弱相关": "文科与理科之间"
    }

    # 成都教育特色建议
    recommendations = [
        {
            "建议": "强化数理化协同教学，构建理科知识体系",
            "数据依据": "PCA显示数理化在第一主成分高度重合，属于同一能力维度",
            "预期效果": "提升成都中学生理科整体水平"
        },
        {
            "建议": "针对文理倾向学生实施分层个性化教学",
            "数据依据": "第二主成分清晰区分文理偏好，学生偏科明显",
            "预期效果": "减少偏科，促进五科均衡发展"
        }
    ]

    insights = {
        "primary_factor": primary_factor,
        "secondary_factor": secondary_factor,
        "subject_correlation": subject_correlation,
        "recommendations": recommendations
    }

    # ==========================
    # 4. 可视化（3张图：PCA散点图、方差解释率、特征向量热力图）
    # ==========================
    fig = plt.figure(figsize=(18, 5))

    # ----------------------
    # 图1：PCA降维散点图（成都区域着色）
    # ----------------------
    ax1 = plt.subplot(1, 3, 1)
    district_colors = {
        '锦江区': '#FF4757',
        '青羊区': '#00D2D3',
        '高新区': '#FF6B35',
        '武侯区': '#78E08F'
    }
    for district, color in district_colors.items():
        mask = df['所在区'] == district
        ax1.scatter(pca_result[mask, 0], pca_result[mask, 1], 
                    s=25, alpha=0.6, color=color, label=district)
    ax1.set_xlabel(f'PC1（{explained_variance[0]:.1%}）', fontsize=12)
    ax1.set_ylabel(f'PC2（{explained_variance[1]:.1%}）', fontsize=12)
    ax1.set_title('成都中学生成绩 PCA 降维分布图', fontsize=13)
    ax1.legend()
    ax1.grid(alpha=0.3)

    # ----------------------
    # 图2：方差解释率条形图
    # ----------------------
    ax2 = plt.subplot(1, 3, 2)
    ax2.bar(['PC1', 'PC2'], explained_variance, color=['#FF6B6B', '#4ECDC4'])
    ax2.set_title('主成分方差解释率', fontsize=13)
    ax2.set_ylabel('方差解释比例', fontsize=12)
    for i, v in enumerate(explained_variance):
        ax2.text(i, v + 0.01, f'{v:.1%}', ha='center', fontweight='bold')

    # ----------------------
    # 图3：特征向量热力图（学科权重）
    # ----------------------
    ax3 = plt.subplot(1, 3, 3)
    im = ax3.imshow(components, cmap='RdBu_r', vmin=-1, vmax=1)
    ax3.set_xticks(range(len(score_columns)))
    ax3.set_yticks([0, 1])
    ax3.set_xticklabels(score_columns, fontsize=11)
    ax3.set_yticklabels(['PC1', 'PC2'], fontsize=11)
    ax3.set_title('学科 - 主成分权重热力图', fontsize=13)
    for i in range(2):
        for j in range(len(score_columns)):
            ax3.text(j, i, f'{components[i,j]:.2f}', 
                     ha='center', va='center', color='white', fontweight='bold')

    plt.tight_layout()
    plt.show()

    # ==========================
    # 5. 按要求输出返回值
    # ==========================
    return pca, explained_variance, components, insights
df, score_columns, schools_info = create_chengdu_education_dataset()
pca_model, explained_variance, components, insights = challenge_1_pca_framework_new(df, score_columns)
```

```python
#挑战2：线性方程组应用 - 教育资源优化模型【2】
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import linprog

def challenge_2_linear_programming_framework():
    """
    挑战2：线性方程组应用 - 教育资源优化模型
    成都教育资源分配线性规划最优解
    返回：严格按要求的字典格式
    """
    plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB']
    plt.rcParams['axes.unicode_minus'] = False

    # ======================
    # 步骤1：参数定义（固定随机种子）
    # ======================
    np.random.seed(42)
    school_names = ['成都七中', '树德中学', '石室中学', '成都外国语学校', '玉林中学']
    n_schools = len(school_names)

    resource_cost = np.random.uniform(0.8, 1.2, n_schools)
    min_investment = np.random.uniform(5, 15, n_schools)
    max_investment = np.random.uniform(20, 40, n_schools)
    improvement_coef = np.random.uniform(0.5, 1.5, n_schools)
    improvement_const = np.random.uniform(0, 5, n_schools)

    # ======================
    # 步骤2：线性规划模型
    # ======================
    c = -improvement_coef  # linprog 默认最小化，最大化要加负号
    A_budget = resource_cost.reshape(1, -1)
    b_budget = np.array([100])
    bounds = [(m, M) for m, M in zip(min_investment, max_investment)]

    # ======================
    # 步骤3：求解模型
    # ======================
    res = linprog(c, A_ub=A_budget, b_ub=b_budget, bounds=bounds, method='highs')

    if not res.success:
        return {"error": res.message}

    optimal_x = res.x
    total_improvement = -res.fun + improvement_const.sum()
    budget_used = np.sum(resource_cost * optimal_x)
    budget_utilization = budget_used / 100

    # ======================
    # 步骤4：约束验证
    # ======================
    feasible = True
    for i in range(n_schools):
        if not (min_investment[i] <= optimal_x[i] <= max_investment[i]):
            feasible = False

    # ======================
    # 步骤5：敏感度分析（预算 ±20%）
    # ======================
    budgets = np.linspace(80, 120, 9)
    sens = {}
    for b in budgets:
        r = linprog(c, A_ub=A_budget, b_ub=[b], bounds=bounds, method='highs')
        if r.success:
            sens[round(b)] = round(-r.fun + improvement_const.sum(), 2)
        else:
            sens[round(b)] = None

    # ======================
    # 步骤6：公平性分析（基尼系数简化版：方差评估）
    # ======================
    allocation_list = list(optimal_x)
    fairness = {
        "mean_alloc": round(float(np.mean(allocation_list)), 2),
        "std_alloc": round(float(np.std(allocation_list)), 2),
        "min_alloc": round(float(np.min(allocation_list)), 2),
        "max_alloc": round(float(np.max(allocation_list)), 2),
        "fairness_level": "均衡" if np.std(allocation_list) < 3 else "不均衡"
    }

    # ======================
    # 步骤7：可视化（敏感度曲线 + 最优分配图）
    # ======================
    plt.figure(figsize=(14, 5))

    # 左图：敏感度分析
    plt.subplot(1, 2, 1)
    plt.plot(list(sens.keys()), list(sens.values()), 'bo-', linewidth=2)
    plt.xlabel("预算（万元）")
    plt.ylabel("总成绩提升（分）")
    plt.title("教育预算敏感度分析")
    plt.grid(alpha=0.3)

    # 右图：最优投资分配
    plt.subplot(1, 2, 2)
    plt.bar(school_names, optimal_x, color='#FF6B6B')
    plt.xticks(rotation=15, fontsize=9)
    plt.title("各校最优教育资源分配（万元）")
    plt.tight_layout()
    plt.show()

    # ======================
    # 输出（严格按要求格式）
    # ======================
    return {
        "optimal_allocation": {school_names[i]: round(optimal_x[i], 2) for i in range(5)},
        "total_improvement": round(total_improvement, 2),
        "budget_utilization": round(budget_utilization, 2),
        "sensitivity": sens,
        "fairness_analysis": fairness
    }
# 直接运行
result = challenge_2_linear_programming_framework()

# 查看结果
import json
print(json.dumps(result, indent=2, ensure_ascii=False))
```

```python
#挑战3：矩阵分解实战 - 智能教育推荐系统【1】
def challenge_3_recommendation_system_framework():
    """
    挑战3：智能教育推荐系统详细框架
    
    输出：
    dict: 推荐系统实现框架和评估指南
    """
    print("\n🎯 挑战3：智能教育推荐系统详细框架")
    print("-" * 60)
    
    print("任务背景:")
    print("开发成都教育科技平台的个性化推荐系统，使用SVD分析学生-学习资源交互矩阵。")
    print("学生200人（5所学校各40人），学习资源50个（视频、习题、电子书、模拟考试）。")
    
    print("\n数据规格:")
    data_spec = """
- 学生数量: 200 (成都5校各40人)
- 学习资源: 50个
  - 视频课程: 20个 (编号R001-R020)
  - 习题集: 15个 (编号R021-R035)
  - 电子书: 10个 (编号R036-R045)
  - 模拟考试: 5个 (编号R046-R050)
- 评分范围: 1-5分 (整数)
- 稀疏度: 80%缺失 (模拟真实场景)
- 资源元数据: 学科标签、难度等级、时长估计、成都课程标准匹配度
    """
    print(data_spec)
    
    print("\n详细实现步骤:")
    
    steps = [
        "# 步骤1: 数据生成与模拟",
        "import numpy as np",
        "from scipy.sparse.linalg import svds",
        "from sklearn.metrics import mean_squared_error",
        "",
        "# 参数设置",
        "n_students = 200",
        "n_resources = 50",
        "sparsity = 0.8  # 80%缺失",
        "",
        "# 初始化评分矩阵",
        "ratings = np.zeros((n_students, n_resources))",
        "",
        "# 生成学生能力水平（正态分布）",
        "np.random.seed(42)",
        "student_ability = np.random.normal(0.5, 0.2, n_students)",
        "student_ability = np.clip(student_ability, 0.1, 0.9)",
        "",
        "# 生成资源难度",
        "resource_difficulty = np.random.normal(0.5, 0.2, n_resources)",
        "resource_difficulty = np.clip(resource_difficulty, 0.1, 0.9)",
        "",
        "# 生成交互评分（考虑能力-难度匹配）",
        "n_ratings = int(n_students * n_resources * (1 - sparsity))",
        "student_indices = np.random.randint(0, n_students, n_ratings)",
        "resource_indices = np.random.randint(0, n_resources, n_ratings)",
        "",
        "# 评分生成逻辑：能力匹配度越高，评分越高",
        "for i, j in zip(student_indices, resource_indices):",
        "    ability = student_ability[i]",
        "    difficulty = resource_difficulty[j]",
        "    ",
        "    # 匹配度计算",
        "    match_score = 1.0 - abs(ability - difficulty)",
        "    ",
        "    # 加入随机性",
        "    noise = np.random.normal(0, 0.2)",
        "    raw_rating = match_score * 4 + 1 + noise  # 映射到1-5分",
        "    rating = int(np.clip(np.round(raw_rating), 1, 5))",
        "    ",
        "    ratings[i, j] = rating",
        "",
        "# 步骤2: SVD矩阵分解",
        "# 选择潜在特征维度",
        "k = 10",
        "",
        "# 执行SVD（处理稀疏矩阵）",
        "U, sigma, Vt = svds(ratings, k=k)",
        "",
        "# 重构矩阵",
        "sigma_matrix = np.diag(sigma)",
        "ratings_pred = U @ sigma_matrix @ Vt",
        "",
        "# 步骤3: 推荐系统评估",
        "# 计算RMSE（仅在有评分的部分）",
        "mask = ratings > 0",
        "rmse = np.sqrt(mean_squared_error(ratings[mask], ratings_pred[mask]))",
        "print(f'RMSE: {rmse:.4f}')",
        "",
        "# 计算MAE",
        "mae = np.mean(np.abs(ratings[mask] - ratings_pred[mask]))",
        "print(f'MAE: {mae:.4f}')",
        "",
        "# 步骤4: 个性化推荐生成",
        "# 为特定学生生成推荐",
        "def generate_recommendations(student_id, ratings, ratings_pred, top_n=5):",
        "    \"\"\"为指定学生生成top-N推荐\"\"\"",
        "    student_ratings = ratings[student_id]",
        "    student_predictions = ratings_pred[student_id]",
        "    ",
        "    # 找出未评分但预测评分高的资源",
        "    unrated_mask = student_ratings == 0",
        "    unrated_indices = np.where(unrated_mask)[0]",
        "    ",
        "    if len(unrated_indices) == 0:",
        "        return []",
        "    ",
        "    # 获取预测评分",
        "    unrated_predictions = student_predictions[unrated_indices]",
        "    ",
        "    # 按预测评分排序",
        "    sorted_indices = np.argsort(unrated_predictions)[::-1]",
        "    top_indices = unrated_indices[sorted_indices[:top_n]]",
        "    ",
        "    return top_indices",
        "",
        "# 示例：为第0号学生生成推荐",
        "student_id = 0",
        "recommendations = generate_recommendations(student_id, ratings, ratings_pred)",
        "print(f'学生{student_id}的top-5推荐资源索引: {recommendations}')",
        "",
        "# 步骤5: 推荐质量分析",
        "# 评估推荐多样性、新颖性、教育价值",
        "def analyze_recommendation_quality(recommendations, resource_metadata):",
        "    \"\"\"分析推荐列表的质量\"\"\"",
        "    quality_metrics = {}",
        "    ",
        "    # 多样性：推荐资源类型的分布",
        "    resource_types = resource_metadata['type']",
        "    rec_types = resource_types[recommendations]",
        "    unique_types = np.unique(rec_types)",
        "    diversity = len(unique_types) / len(np.unique(resource_types))",
        "    ",
        "    quality_metrics['diversity'] = diversity",
        "    ",
        "    # 新颖性：推荐资源的热门程度（模拟）",
        "    # 假设热门资源被更多学生评分",
        "    resource_popularity = np.sum(ratings > 0, axis=0)",
        "    rec_popularity = resource_popularity[recommendations]",
        "    novelty = 1.0 - np.mean(rec_popularity) / np.max(resource_popularity)",
        "    ",
        "    quality_metrics['novelty'] = novelty",
        "    ",
        "    # 教育价值：难度匹配度和知识覆盖度",
        "    # 这里需要根据具体业务逻辑实现",
        "    ",
        "    return quality_metrics",
        "",
        "# 步骤6: 成都特色融入",
        "# 资源标注成都课程标准匹配度",
        "# 考虑区域教育资源差异",
        "# 推荐时兼顾升学政策导向",
    ]
    
    for step in steps:
        print(step)
    
    print("\n预期输出:")
    expected_output = {
        'data_generation': {
            'ratings_matrix': '200×50评分矩阵',
            'student_ability': '200维学生能力数组',
            'resource_difficulty': '50维资源难度数组',
            'sparsity': '实际缺失率'
        },
        'svd_results': {
            'U_matrix': '200×10左奇异矩阵',
            'sigma': '10维奇异值数组',
            'Vt_matrix': '10×50右奇异矩阵转置',
            'reconstructed_matrix': '200×50重构评分矩阵'
        },
        'evaluation_metrics': {
            'RMSE': '均方根误差',
            'MAE': '平均绝对误差',
            'precision_at_k': '前k推荐准确率'
        },
        'recommendation_output': {
            'top_recommendations': '为每位学生的推荐列表',
            'quality_analysis': '推荐质量评估结果',
            'chengdu_features': '成都特色融入分析'
        }
    }
    
    for key, value in expected_output.items():
        print(f"- {key}: {value}")
    
    return expected_output
challenge_3_recommendation_system_framework()
```

```python
#挑战3：矩阵分解实战 - 智能教育推荐系统【2】
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.sparse.linalg import svds
from sklearn.metrics import mean_squared_error

def challenge_3_recommendation_system():
    """
    挑战3：矩阵分解实战 - 成都智能教育推荐系统
    功能：生成稀疏评分矩阵 + SVD分解 + 推荐生成 + 评估 + 可视化
    """
    # 中文显示
    plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB']
    plt.rcParams['axes.unicode_minus'] = False

    # ==========================================
    # 步骤1：参数设置（严格按任务要求）
    # ==========================================
    np.random.seed(42)
    n_students = 200    # 5所学校 × 40人
    n_resources = 50    # 视频20 + 习题15 + 电子书10 + 模拟考试5
    sparsity = 0.8      # 80%缺失
    k = 10              # 隐特征维度

    # 学校标签（成都5校）
    school_names = ['成都七中', '树德中学', '石室中学', '成都外国语学校', '玉林中学']
    student_school = np.array([school_names[i//40] for i in range(n_students)])

    # 资源类型标签
    resource_types = (
        ['视频课程']*20 + ['习题集']*15 + ['电子书']*10 + ['模拟考试']*5
    )
    resource_ids = [f'R{i+1:03d}' for i in range(n_resources)]

    # ==========================================
    # 步骤2：生成学生能力 & 资源难度
    # ==========================================
    student_ability = np.random.normal(0.5, 0.2, n_students)
    student_ability = np.clip(student_ability, 0.1, 0.9)

    resource_difficulty = np.random.normal(0.5, 0.2, n_resources)
    resource_difficulty = np.clip(resource_difficulty, 0.1, 0.9)

    # ==========================================
    # 步骤3：生成稀疏评分矩阵（1-5分，20%密度）
    # ==========================================
    ratings = np.zeros((n_students, n_resources))
    n_ratings = int(n_students * n_resources * (1 - sparsity))

    student_idx = np.random.randint(0, n_students, n_ratings)
    resource_idx = np.random.randint(0, n_resources, n_ratings)

    for i, j in zip(student_idx, resource_idx):
        match = 1 - abs(student_ability[i] - resource_difficulty[j])
        noise = np.random.normal(0, 0.2)
        raw = match * 4 + 1 + noise
        ratings[i, j] = int(np.clip(np.round(raw), 1, 5))

    # ==========================================
    # 步骤4：SVD 矩阵分解
    # ==========================================
    U, sigma, Vt = svds(ratings, k=k)
    sigma_matrix = np.diag(sigma)
    ratings_pred = U @ sigma_matrix @ Vt

    # ==========================================
    # 步骤5：模型评估（RMSE, MAE）
    # ==========================================
    mask = ratings > 0
    rmse = np.sqrt(mean_squared_error(ratings[mask], ratings_pred[mask]))
    mae = np.mean(np.abs(ratings[mask] - ratings_pred[mask]))

    # ==========================================
    # 步骤6：推荐生成函数（Top-5 未评分资源）
    # ==========================================
    def get_topN_recommend(student_id, top_n=5):
        scored = ratings[student_id] > 0
        unrated = ~scored
        unrated_idx = np.where(unrated)[0]
        pred_scores = ratings_pred[student_id, unrated_idx]
        top_idx = unrated_idx[np.argsort(pred_scores)[::-1][:top_n]]
        return top_idx

    # ==========================================
    # 步骤7：推荐多样性 & 教育价值分析
    # ==========================================
    def analyze_quality(rec_list):
        types = [resource_types[i] for i in rec_list]
        diversity = len(set(types)) / 4  # 共4类资源
        popularity = np.sum(ratings[:, rec_list] > 0, axis=0)
        novelty = 1 - np.mean(popularity) / np.max(ratings > 0)
        return {
            "diversity": round(diversity, 2),
            "novelty": round(novelty, 2),
            "recommend_types": types
        }

    # ==========================================
    # 步骤8：为所有学生生成推荐
    # ==========================================
    all_recommendations = {}
    quality_report = {}
    for sid in range(n_students):
        recs = get_topN_recommend(sid, 5)
        all_recommendations[f"学生{sid}"] = [resource_ids[i] for i in recs]
        quality_report[f"学生{sid}"] = analyze_quality(recs)

    # ==========================================
    # 步骤9：成都特色分析（按学校）
    # ==========================================
    chengdu_analysis = {}
    for school in school_names:
        sids = np.where(student_school == school)[0]
        school_recs = [get_topN_recommend(s, 5) for s in sids]
        all_types = []
        for rs in school_recs:
            all_types.extend([resource_types[r] for r in rs])
        chengdu_analysis[school] = pd.Series(all_types).value_counts().to_dict()

    # ==========================================
    # 步骤10：可视化
    # ==========================================
    plt.figure(figsize=(16,5))

    # 图1：真实 vs 预测评分分布
    plt.subplot(1,3,1)
    plt.hist(ratings[mask], bins=5, alpha=0.5, label='真实评分')
    plt.hist(ratings_pred[mask], bins=20, alpha=0.5, label='预测评分')
    plt.title('评分分布对比')
    plt.legend()

    # 图2：奇异值分布
    plt.subplot(1,3,2)
    plt.bar(range(1,11), sigma[::-1], color='orange')
    plt.title('SVD奇异值分布')

    # 图3：成都各校推荐类型分布
    plt.subplot(1,3,3)
    pd.DataFrame(chengdu_analysis).T.plot(kind='bar', ax=plt.gca())
    plt.title('成都各校推荐资源类型分布')
    plt.xticks(rotation=15)

    plt.tight_layout()
    plt.show()

    # ==========================================
    # 输出结果（严格按要求）
    # ==========================================
    return {
        "data": {
            "ratings_matrix": ratings,
            "sparsity": round(1 - mask.mean(), 2),
            "student_school": student_school,
            "resource_types": resource_types
        },
        "svd": {
            "U": U, "sigma": sigma, "Vt": Vt, "ratings_pred": ratings_pred
        },
        "evaluation": {
            "RMSE": round(rmse, 4),
            "MAE": round(mae, 4)
        },
        "recommendations": all_recommendations,
        "quality_report": quality_report,
        "chengdu_feature_analysis": chengdu_analysis
    }
# 直接调用
result = challenge_3_recommendation_system()
# 查看评估指标
print("RMSE:", result['evaluation']['RMSE'])
print("MAE:", result['evaluation']['MAE'])

# 查看学生0推荐
print("学生0推荐:", result['recommendations']['学生0'])

# 查看成都各校推荐偏好
print("\n成都各校推荐分布：")
for school, cnt in result['chengdu_feature_analysis'].items():
    print(school, cnt)
```
