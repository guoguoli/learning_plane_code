```python
import numpy as np
import matplotlib.pyplot as plt

# 创建单位正方形顶点（二维向量集合）
square = np.array([[0, 1, 1, 0, 0],  # x坐标
                   [0, 0, 1, 1, 0]]) # y坐标

# 定义复合线性变换：先旋转45°，再非均匀缩放
theta = np.pi / 4  # 45度弧度制
rotation = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta), np.cos(theta)]])
scaling = np.array([[1.5, 0],   # x轴缩放1.5倍
                    [0, 0.8]])  # y轴缩放0.8倍

# 矩阵乘法表示变换复合：先旋转后缩放
transform = scaling @ rotation  # 注意矩阵乘法顺序

# 应用变换：矩阵乘以每个向量
transformed = transform @ square

# 可视化对比
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 原始正方形
axes[0].plot(square[0], square[1], 'b-', linewidth=2)
axes[0].fill(square[0], square[1], 'blue', alpha=0.2)
axes[0].set_title('原始单位正方形', fontsize=14)
axes[0].set_xlabel('x轴')
axes[0].set_ylabel('y轴')
axes[0].grid(True, alpha=0.3)
axes[0].axis('equal')

# 变换后的平行四边形
axes[1].plot(transformed[0], transformed[1], 'r-', linewidth=2)
axes[1].fill(transformed[0], transformed[1], 'red', alpha=0.2)
axes[1].set_title('线性变换后：旋转45°+非均匀缩放', fontsize=14)
axes[1].set_xlabel('x轴')
axes[1].set_ylabel('y轴')
axes[1].grid(True, alpha=0.3)
axes[1].axis('equal')

plt.tight_layout()
plt.show()

# 数学验证：变换矩阵的行列式值
det_value = np.linalg.det(transform)
print(f"变换矩阵的行列式值: {det_value:.4f}")
print(f"几何解释：面积变为原来的 {det_value:.2f} 倍")
```

```python
#numpy 高级线性代数实战
import numpy as np
print("🔢 NumPy高级线性代数实战")
print("=" * 60)

# 1. 矩阵乘法
# 矩阵乘法是线性代数中最常用的运算符，它将两个矩阵相乘，得到一个新的矩阵。
# 矩阵乘法的运算规则是：两个矩阵的行数等于第一个矩阵的列数，第二个矩阵的列数等于第二个矩阵的行数。
# 矩阵乘法的结果是一个新的矩阵，其行数等于第一个矩阵的行数，列数等于第二个矩阵的列数。
# 矩阵乘法的运算结果可以用一个公式来表示：
# C = A @ B
# 其中，C是乘积矩阵，A和B是两个矩阵。  
# 矩阵乘法的运算结果是一个新的矩阵，其元素是两个矩阵对应位置元素的乘积。
#代码示列
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
arr3 = arr1 @ arr2 
print(arr3)
# 2. 矩阵的转置
# 矩阵的转置是指矩阵的行列互换，即将矩阵的行变为列，将矩阵的列变为行。
# 矩阵的转置运算结果是一个新的矩阵，其行数等于原矩阵的列数，列数等于原矩阵的行数。
# 矩阵的转置运算可以用一个公式来表示：
# C = A.T
# 其中，C是转置矩阵，A是矩阵。
# 代码示列
arr4 = np.array([[1, 2], [3, 4]])
arr5 = arr4.T
print(arr5)
# 3. 矩阵的逆
# 矩阵的逆是指矩阵的乘法逆元，即矩阵乘以其转置矩阵，得到的矩阵是单位矩阵。
# 矩阵的逆运算结果是一个新的矩阵，其行数等于原矩阵的列数，列数等于原矩阵的行数。
# 矩阵的逆运算可以用一个公式来表示：
# C = np.linalg.inv(A)
# 其中，C是逆矩阵，A是矩阵。
# 代码示列
arr6 = np.array([[1, 2], [3, 4]])
arr7 = np.linalg.inv(arr6)
print(arr7)
# 4. 矩阵的行列式
# 矩阵的行列式是指矩阵的乘积的第n行第n列元素，其中n是矩阵的阶数。  
# 矩阵的行列式的值是矩阵的秩。
# 矩阵的行列式运算可以用一个公式来表示：
# det(A) = |A|
# 其中，det(A)是矩阵A的行列式，|A|是矩阵A的秩
# 代码示列
arr8 = np.array([[1, 2], [3, 4]])
arr9 = np.linalg.det(arr8)
print(arr9)
# 5. 矩阵的秩   
# 矩阵的秩是指矩阵的行列式的绝对值。
# 矩阵的秩的值是矩阵的行数与列数的最大值。
# 矩阵的秩运算可以用一个公式来表示：
# rank(A) = r
# 其中，rank(A)是矩阵A的秩，r是矩阵A的行数与列数的最大值。
# 代码示列
arr10 = np.array([[1, 2], [3, 4]])
arr11 = np.linalg.matrix_rank(arr10)
print(arr11)
# 6. 矩阵的特征值和特征向量
# 矩阵的特征值和特征向量是指矩阵的一些特殊的数值。
# 矩阵的特征值是指矩阵的一些特殊的数值，这些数值是矩阵的一些线性变换的结果。
# 矩阵的特征向量是指矩阵的特征值对应的单位向量。   
# 矩阵的特征值和特征向量运算可以用两个公式来表示：
# eigenvalues, eigenvectors = np.linalg.eig(A)
# 其中，eigenvalues是矩阵A的特征值，eigenvectors是矩阵A的特征向量。
# 代码示列
arr12 = np.array([[1, 2], [3, 4]])
arr13 = np.linalg.eig(arr12)
print(arr13)
# 7. 矩阵的条件数
# 矩阵的条件数是指矩阵的秩与矩阵的最大特征值的比值。
# 矩阵的条件数的值越大，矩阵越不稳定。
# 矩阵的条件数运算可以用一个公式来表示：
# cond(A) = ||A|| / ||A^2||
# 其中，cond(A)是矩阵A的条件数，||A||是矩阵A的秩，||A^2||是矩阵A的最大特征值的平方根。
# 代码示列
arr14 = np.array([[1, 2], [3, 4]])
arr15 = np.linalg.cond(arr14)
print(arr15)
#特殊矩阵的创建于性质验证
print("=" * 60)
print("🔢 特殊矩阵的创建与性质验证")
print("=" * 60)
# 1. 单位矩阵
# 单位矩阵是指对角线元素为1，其他元素为0的矩阵。
# 单位矩阵的创建可以用一个公式来表示：
# I = np.eye(n) 
#示列代码
i1 = np.eye(3)
print(i1)
# 2. 对称矩阵
# 对称矩阵是指对称的矩阵，即矩阵的转置等于矩阵本身。    
# 对称矩阵的创建可以用一个公式来表示：
# S = np.array([[a, b], [b, c]])
# 其中，S是对称矩阵，a, b, c是任意实数。
# 示列代码
s1 = np.array([[1, 2], [2, 1]])
print(s1)
# 3. 正定矩阵
# 正定矩阵是指矩阵的每个特征值都大于零的矩阵。
# 正定矩阵的创建可以用一个公式来表示：
# P = np.array([[a, b], [b, c]])
# 其中，P是正定矩阵，a, b, c是任意实数。
# 示列代码
p1 = np.array([[2, 1], [1, 2]])
print(p1)
# 4. 负定矩阵
# 负定矩阵是指矩阵的每个特征值都小于零的矩阵。
# 负定矩阵的创建可以用一个公式来表示：
# N = np.array([[a, b], [b, c]])
# 其中，N是负定矩阵，a, b, c是任意实数。
# 示列代码
n1 = np.array([[-2, 1], [1, -2]])
print(n1)
# 5. 奇异矩阵
# 奇异矩阵是指矩阵的行列式为0的矩阵。
# 奇异矩阵的创建可以用一个公式来表示：
# O = np.array([[a, b], [b, c]])
# 其中，O是奇异矩阵，a, b, c是任意实数。
# 示列代码
o1 = np.array([[1, 2], [2, 1]])
print(o1)
# 6. 行列式为0的矩阵
# 行列式为0的矩阵是指矩阵的行列式为0的矩阵。
# 行列式为0的矩阵的创建可以用一个公式来表示：
# Z = np.array([[a, b], [b, c]])
# 其中，Z是行列式为0的矩阵，a, b, c是任意实数。
# 示列代码
z1 = np.array([[1, 2], [2, 1]])
print(z1)
# 7. 秩为奇数的矩阵
# 秩为奇数的矩阵是指矩阵的秩为奇数的矩阵。
# 秩为奇数的矩阵的创建可以用一个公式来表示：
# R = np.array([[a, b], [b, c]])
# 其中，R是秩为奇数的矩阵，a, b, c是任意实数。
# 示列代码
r1 = np.array([[1, 2, 3], [2, 1, 4], [3, 4, 1]])
print(r1)
# 8. 秩为偶数的矩阵
# 秩为偶数的矩阵是指矩阵的秩为偶数的矩阵。
# 秩为偶数的矩阵的创建可以用一个公式来表示：
# E = np.array([[a, b], [b, c]])
# 其中，E是秩为偶数的矩阵，a, b, c是任意实数。
# 示列代码
e1 = np.array([[1, 2, 3], [2, 1, 4], [3, 4, 1]])
print(e1)  
# 9. 正交矩阵
# 正交矩阵是指矩阵的每一列都是单位向量的矩阵。
# 正交矩阵的创建可以用一个公式来表示：
# Q = np.array([[a, b], [b, c]])
# 其中，Q是正交矩阵，a, b, c是任意实数。
# 示列代码
q1 = np.array([[1, 0], [0, 1]])
print(q1)
np.random.seed(42)
a = np.random.rand(4, 4)
q,r = np.linalg.qr(a)
print(f"\n随机正交矩阵q (形状{q.shape}):")
print(f"前2行:\n{q[:2]}")
print(f"正交性验证:qᵀq≈ i1? (最大误差:{np.max(np.abs(q.T @ q - np.eye(4))):.6f})")
# 10. 酉矩阵
# 酉矩阵是指矩阵的每一列都是单位向量，且矩阵的转置等于其逆矩阵的矩阵。
# 酉矩阵的创建可以用一个公式来表示：
# H = np.array([[a, b], [b, c]])
# 其中，H是酉矩阵，a, b, c是任意实数。
# 示列代码
h1 = np.array([[1, 0], [0, 1]]) 
print(h1)
# 11. 矩阵分解的数学意义与实现
print("\n2. 矩阵分解的核心应用:")
matrix = np.array([[4, 2], [2, 3]])  # 对称正定矩阵
print(f"对称正定矩阵A:\n{matrix}")
# 特征值分解：A = VΛV⁻¹
eigenvalues, eigenvectors = np.linalg.eig(matrix)  # 特征值分解
print(f"\n特征值:\n{eigenvalues}")
print(f"特征向量矩阵 V:\n{eigenvectors}")
print(f"验证: ||AV - VΛ||_F = {np.linalg.norm(matrix @ eigenvectors - eigenvectors @ np.diag(eigenvalues)):.6f}")
# 奇异值分解：A = UΣVᵀ
u, s, vt = np.linalg.svd(matrix)  # 奇异值分解
print(f"\n奇异值:\n{s}")
print(f"左奇异矩阵 U:\n{u}")
print(f"右奇异矩阵 Vᵀ:\n{vt}")
print(f"Σ矩阵:\n{np.diag(s)}")
print(f"重构误差: ||A - UΣVᵀ||_F = {np.linalg.norm(matrix - u @ np.diag(s) @ vt):.6f}")
# 线性方程组求解的数值稳定性分析
print("\n3. 线性方程组求解的数值稳定性分析:")
a2 = np.array([[3,2],[1,-1]])
b2 = np.array([5,1])
cond_num = np.linalg.cond(a2)
print(f"系数矩阵条件数: κ(A) = {cond_num:.4f}")
print(f"解释: 条件数较大(>10³)时，小扰动可能导致解大幅变化")
x = np.linalg.solve(a2, b2)
print(f"方程组的解: x = {x[0]:.2f}, y = {x[1]:.2f}")
print(f"残差范数: ||Ax - b||₂ = {np.linalg.norm(a2 @ x - b2):.6f}")
```

```python
 #Pandas数据处理实战
import pandas as pd

np.random.seed(42)
schools_info = {
        '成都七中': {'district': '锦江区', 'tier': '顶尖', 'student_count': 3000},
        '树德中学': {'district': '青羊区', 'tier': '顶尖', 'student_count': 2800},
        '石室中学': {'district': '青羊区', 'tier': '顶尖', 'student_count': 3200},
        '成都外国语学校': {'district': '高新区', 'tier': '一流', 'student_count': 2500},
        '玉林中学': {'district': '武侯区', 'tier': '一流', 'student_count': 2000}
    }
n_students = 1000
schools = list(schools_info.keys())
districts = [schools_info[school]['district'] for school in schools]
tiers = [schools_info[school]['tier'] for school in schools]
student_counts = [schools_info[school]['student_count'] for school in schools]

school_probs = np.array(student_counts) / np.sum(student_counts)
print(school_probs)
print(f"1. 成都教育资源分布:")
for school, info in schools_info.items():
    print(f"     - {school}: {info['district']}区， {info['tier']}，约{info['student_count']}名学生")

data = {
    '学生ID':range(1,n_students+1),
    '姓名':[f'学生_{i:04d}' for i in range(1,n_students+1)],
    '性别':np.random.choice(['男', '女'], n_students,p=[0.52, 0.48]),
    '年龄':np.random.choice([15,16,17,18],n_students,p=[0.25,0.35,0.30,0.10]),

    '学校':np.random.choice(schools, n_students, p=school_probs),
    '所在区':np.random.choice(districts, n_students),
    '学校层级':np.random.choice(tiers, n_students),

    '语文':np.zeros(n_students),
    '数学':np.zeros(n_students),
    '英语':np.zeros(n_students),
    '物理':np.zeros(n_students),
    '化学':np.zeros(n_students),

    '每周学习时长(小时)':np.zeros(n_students),
    '每周辅导班时长(小时)':np.zeros(n_students),
    '教育APP使用指数':np.zeros(n_students),

    '学习压力指数': np.zeros(n_students),
    '睡眠质量评分': np.zeros(n_students),
    '课外活动参与度': np.zeros(n_students),
    
}
for i, school in enumerate(schools):
    mask = data['学校'] == school
    n_school_students = np.sum(mask)
    if n_school_students == 0:
        continue

    if schools_info[school]['tier'] == '顶尖':
        base_mean = 75
        base_std = 10
        pressure_mean = 70
    else:
        base_mean = 65
        base_std = 15
        pressure_mean = 60
    # 生成相关学科成绩
    # 语文-数学相关系数约0.6，数学-物理约0.7
    cov_matrix = np.array([
        [1.0, 0.6, 0.5, 0.4, 0.3],  # 语文
        [0.6, 1.0, 0.7, 0.6, 0.5],  # 数学
        [0.5, 0.7, 1.0, 0.5, 0.4],  # 英语
        [0.4, 0.6, 0.5, 1.0, 0.8],  # 物理
        [0.3, 0.5, 0.4, 0.8, 1.0]   # 化学
    ])
    # 生成多元正态分布成绩
    means = np.array([base_mean,base_mean+3,base_mean-2,base_mean-5,base_mean-7])
    scores = np.random.multivariate_normal(means, cov_matrix * base_std**2, n_school_students)
    data['语文'][mask] = scores[:,0].clip(0,100)
    data['数学'][mask] = scores[:,1].clip(0,100)
    data['英语'][mask] = scores[:,2].clip(0,100)
    data['物理'][mask] = scores[:,3].clip(0,100)
    data['化学'][mask] = scores[:,4].clip(0,100)
    # 生成行为数据（与成绩相关）
    data['每周学习时长(小时)'][mask] = np.random.normal(30, 8, n_school_students).clip(10, 50)
    data['每周辅导班时长(小时)'][mask] = np.random.normal(8, 4, n_school_students).clip(0, 20)
    data['教育APP使用指数'][mask] = np.random.normal(7, 3, n_school_students).clip(0, 15)

    # 心理指标
    data['学习压力指数'][mask] = np.random.normal(pressure_mean, 15, n_school_students).clip(20, 100)
    data['睡眠质量评分'][mask] = np.random.normal(70, 12, n_school_students).clip(30, 100)
    data['课外活动参与度'][mask] = np.random.normal(60, 20, n_school_students).clip(0, 100)
df = pd.DataFrame(data)
print(f"\n2. 数据集概览:")
print(f"   总记录数: {df.shape[0]}名学生")
print(f"   特征维度: {df.shape[1]}个字段")
print(f"\n3. 数据结构:")
print(df.dtypes.to_string())
print(f"\n4. 描述性统计（学科成绩）:")
score_columns = ['语文', '数学', '英语', '物理', '化学']
print(df[score_columns].describe().round(2))    
print(f"\n5. 学校分布:")
school_dist = df['学校'].value_counts()

for school, count in school_dist.items():
    tier = schools_info[school]['tier']
    district = schools_info[school]['district']
    print(f"   - {school}（{district}，{tier}）: {count}人 ({count/n_students*100:.1f}%)")

# 数据清洗与特征工程
print(f"\n6. 数据清洗与特征工程:")
 # 检查缺失值
missing_values = df.isnull().sum()
missing_cols = missing_values[missing_values > 0]
if len(missing_cols) > 0:
    print(f"   缺失值统计:\n{missing_cols}")
else:
    print("   无缺失值 ✓")
# 处理异常值（学科成绩限制在0-100）
for col in score_columns:
    df[col] = df[col].clip(0, 100)
print("   学科成绩异常值处理完成（限制在0-100范围内）")

# 创建衍生特征
print(f"\n7. 衍生特征创建:")
# 学科总分与平均分
df['总分'] = df[score_columns].sum(axis=1)
df['平均分'] = df[score_columns].mean(axis=1)

# 学科优势（最高分学科）
df['优势学科'] = df[score_columns].idxmax(axis=1)

# 文理倾向指数（文科=语文+英语，理科=数学+物理+化学）
df['文科得分'] = df['语文'] + df['英语']
df['理科得分'] = df['数学'] + df['物理'] + df['化学']
df['文理倾向指数'] = (df['文科得分'] - df['理科得分']) / 500  # 归一化到[-1,1]
    
# 学习效率指数 = 平均分 / sqrt(学习时长+辅导班时长)
total_study_hours = df['每周学习时长(小时)'] + df['每周辅导班时长(小时)']
df['学习效率指数'] = df['平均分'] / np.sqrt(total_study_hours)
    
# 综合素质评分（成绩60% + 课外活动20% + 睡眠质量20%）
df['综合素质评分'] = (df['平均分'] * 0.6 + 
                    df['课外活动参与度'] * 0.2 + 
                    df['睡眠质量评分'] * 0.2)
    
print("   衍生特征创建完成，新增字段:")
new_features = ['总分', '平均分',  '文科得分', '理科得分', 
                '文理倾向指数', '学习效率指数', '综合素质评分']
for feat in new_features:
    print(f"     - {feat}: {df[feat].describe()['mean']:.2f} ± {df[feat].describe()['std']:.2f}")
  
```
